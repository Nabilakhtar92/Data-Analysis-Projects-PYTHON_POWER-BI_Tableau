Exploratory Data Analysis

GENDER

```python
ax = sns.countplot(x='Gender',data =df)
for bars in ax.containers:
  ax.bar_label(bars)
```

The above plot shows there are more number of women indulge in shoppings than men.

AGE

```python
df.columns
```

```python
ax = sns.countplot (data= df, x= 'Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

The above plot shows the age gap where the miost number of shopping is done by both the genders which is between twenty-six to thirty-five.

State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head()
sns.set(rc={'figure.figsize':(15 , 5)})
```

```python
sns.barplot(data=sales_state, x="State" ,y="Orders")
```

The above plotshows the amount of orders generating from different states.


```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head()
sns.set(rc={'figure.figsize':(15 , 5)})
```

```python
sns.barplot(data=sales_state, x="State" ,y="Amount")
```

The above plt shows the purchasing power of different states.

```python
ax = sns.countplot (data= df, x= 'Marital_Status', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
  sns.set(rc={'figure.figsize':(6 , 5)})
```

```python
ax = sns.countplot (data= df, x= 'Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
  sns.set(rc={'figure.figsize':(35,5)})
```

Above plt shows the various occcupation of people or buyers ordering the products.

```python
ax = sns.countplot (data= df, x= 'Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
  sns.set(rc={'figure.figsize':(35,15)})
```

The above plot shows the category in which people are ordering items.

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(data=sales_state, x="Product_Category" ,y="Amount")
```

Thias shows the the amount of money spend on different categories
