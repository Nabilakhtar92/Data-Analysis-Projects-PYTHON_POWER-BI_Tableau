```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
```

```python
plt.style.use('dark_background')
```

```python
df=pd.read_csv('/content/cars_ds_final.csv', encoding='unicode_escape')
df.head()
```

```python
df=df.fillna('')
df=df.replace(' ', '')
```

**DATA VISUALISATIUON**






```python
df[df.Make=='Tata'].tail()
```

**THE NAME OF TOP CAR MANUFACTURING COMPANIES IN INDIA**

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.subplots()
df.Make.value_counts().plot(ax=ax, kind='pie', autopct='%1.1f%%')
ax.set_ylabel("")
ax.set_title("TOP CAR MAKING COMPANIES IN INDIAN MARKETS")
plt.show()
```

**CARS CATEGORISED BY BODY TYPE**

```python
plt.figure(figsize=(18,7))
sns.countplot(data=df, y='Body_Type',color='blue')
plt.title('CARS CATEGORISED BY BODY TYPE')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('')
plt.ylabel('');
```

**CARS DIVIDED IN DATA BASED ON FUEL TYPE**

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.subplots()
df.Fuel_Type.value_counts().plot(ax=ax, kind='pie', autopct='%1.1f%%' )
ax.set_ylabel("")
ax.set_title("Cars Count by Engine Fuel Type")
plt.show()
```

**PlOT SHOWING EX SHOWROOM PRICE BASED ON FUEL TYPE AND DISPLACEMENT OF DIFFERENT CAR MANUFATURING COMPANIES**

```python
fig = px.scatter_3d(df, x='Displacement', z='Ex-Showroom_Price', y='Fuel_Type',color='Make',width=800,height=750)
fig.update_layout(showlegend=True)
fig.show();
```
